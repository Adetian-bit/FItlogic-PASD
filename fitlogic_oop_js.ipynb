## JavaScript OOP - `User`
Menyimpan data pengguna di browser dan menghitung BMI, BMI target, serta tujuan diet.

```python
JS_USER = r"""
class User {
  constructor({name, age, gender, height, weight, targetWeight, activity}) {
    this.name = name || "Pengguna";
    this.age = age;
    this.gender = gender;
    this.height = height;
    this.weight = weight;
    this.targetWeight = targetWeight;
    this.activity = activity;
  }

  get heightM() {
    return this.height / 100;
  }

  get bmi() {
    return this.weight / (this.heightM * this.heightM);
  }

  get targetBMI() {
    return this.targetWeight / (this.heightM * this.heightM);
  }

  get direction() {
    if (this.targetWeight < this.weight - 0.5) return "loss";
    if (this.targetWeight > this.weight + 0.5) return "gain";
    return "maintain";
  }
}
"""
```

## JavaScript OOP - `CalorieCalculator`

```python
JS_CALORIE_CALCULATOR = r"""
class CalorieCalculator {
  constructor(user, intensity) {
    this.user = user;
    this.intensity = intensity;
  }

  calculateBMR() {
    const u = this.user;
    if (u.gender === "male") {
      return (10 * u.weight) + (6.25 * u.height) - (5 * u.age) + 5;
    }
    return (10 * u.weight) + (6.25 * u.height) - (5 * u.age) - 161;
  }

  calculateTDEE(bmr) {
    return bmr * this.user.activity;
  }

  getAdjustment() {
    if (this.user.direction === "loss") return this.intensity === "fast" ? -500 : -250;
    if (this.user.direction === "gain") return this.intensity === "gainFast" ? 400 : 250;
    return 0;
  }

  getProgramLabel() {
    if (this.user.direction === "loss") {
      return this.intensity === "fast" ? "Turun Cepat - Defisit 500 kkal" : "Turun Perlahan - Defisit 250 kkal";
    }
    if (this.user.direction === "gain") {
      return this.intensity === "gainFast" ? "Naik Cepat - Surplus 400 kkal" : "Naik Perlahan - Surplus 250 kkal";
    }
    return "Maintain - Pertahankan Berat";
  }

  getDirectionText() {
    if (this.user.direction === "loss") return "menurunkan berat badan";
    if (this.user.direction === "gain") return "menaikkan berat badan";
    return "mempertahankan berat badan";
  }

  calculate() {
    const bmr = this.calculateBMR();
    const tdee = this.calculateTDEE(bmr);
    const delta = this.getAdjustment();
    return {
      bmr: bmr,
      tdee: tdee,
      delta: delta,
      targetCalories: tdee + delta,
      programLabel: this.getProgramLabel(),
      directionText: this.getDirectionText()
    };
  }
}
"""
```

## JavaScript OOP - `MacroRecommendation`
Menghitung protein, karbohidrat, dan lemak.

```python
JS_MACRO_RECOMMENDATION = r"""
class MacroRecommendation {
  constructor(targetCalories, weight, direction) {
    this.targetCalories = targetCalories;
    this.weight = weight;
    this.direction = direction;
  }

  calculate() {
    let proteinFactor = this.direction === "loss" ? 1.8 : this.direction === "gain" ? 1.7 : 1.6;
    let protein = this.weight * proteinFactor;
    protein = Math.min(protein, this.targetCalories * 0.35 / 4);

    let fat = Math.max(this.weight * 0.6, this.targetCalories * 0.23 / 9);
    fat = Math.min(fat, this.targetCalories * 0.35 / 9);

    let carbCalories = this.targetCalories - (protein * 4) - (fat * 9);

    if (carbCalories < this.targetCalories * 0.18) {
      carbCalories = this.targetCalories * 0.22;
      const remaining = this.targetCalories - carbCalories;
      protein = Math.min(protein, remaining * 0.58 / 4);
      fat = Math.max(0, (remaining - protein * 4) / 9);
    }

    return {
      protein: Math.max(0, protein),
      carb: Math.max(0, carbCalories / 4),
      fat: Math.max(0, fat),
      proteinCal: Math.max(0, protein * 4),
      carbCal: Math.max(0, carbCalories),
      fatCal: Math.max(0, fat * 9)
    };
  }
}
"""
```

## JavaScript OOP - `DietProgress`
Menghitung estimasi waktu, timeline, dan grafik tren target berat.

```python
JS_DIET_PROGRESS = r"""
class DietProgress {
  constructor(user, intensity) {
    this.user = user;
    this.intensity = intensity;
  }

  getRate() {
    if (this.user.direction === "loss") return this.intensity === "fast" ? 0.5 : 0.25;
    if (this.user.direction === "gain") return this.intensity === "gainFast" ? 0.35 : 0.25;
    return 0;
  }

  calculate() {
    const absDiff = Math.abs(this.user.targetWeight - this.user.weight);
    const rate = this.getRate();
    return {
      absDiff: absDiff,
      rate: rate,
      estimateWeeks: rate === 0 ? 0 : Math.ceil(absDiff / rate)
    };
  }

  static calculateProgressPercent(result, logs) {
    if (logs && logs.length) {
      const latest = Number(logs[0].weight);
      const total = Math.abs(result.targetWeight - result.weight);
      if (total < 0.1) return 100;
      const moved = Math.abs(latest - result.weight);
      return Math.round(Utils.clamp((moved / total) * 100, 0, 100));
    }
    return 0;
  }

  static makeTimeline(result) {
    if (result.estimateWeeks === 0) {
      return `<div class="step"><div class="week">Maintain</div><b>Fokus stabil</b><p class="help">Pertahankan berat sekitar ${Utils.one(result.weight)} kg dan pantau setiap minggu.</p></div>`;
    }

    const steps = [0, 0.25, 0.5, 0.75, 1];
    return steps.map((progress, index) => {
      const week = Math.round(result.estimateWeeks * progress);
      const value = result.weight + (result.targetWeight - result.weight) * progress;
      const title = index === 0 ? "Mulai" : index === 4 ? "Target" : "Tahap " + index;
      return `<div class="step"><div class="week">Minggu ${week}</div><b>${title}</b><p class="help">Perkiraan berat: <b>${Utils.one(value)} kg</b></p></div>`;
    }).join("");
  }

  static makeTrendSvg(result, logs) {
    const width = 900, height = 300, padding = 46;
    let points = [];

    if (logs && logs.length >= 2) {
      points = [...logs].reverse().map(log => ({label: log.date.slice(5), weight: Number(log.weight)}));
    } else {
      const weeks = Math.max(4, Math.min(12, result.estimateWeeks || 4));
      for (let i = 0; i <= weeks; i++) {
        points.push({label: "M" + i, weight: result.weight + (result.targetWeight - result.weight) * (i / weeks)});
      }
    }

    points.push({label: "Target", weight: result.targetWeight, target: true});

    const values = points.map(p => p.weight);
    let minValue = Math.min(...values), maxValue = Math.max(...values);
    const margin = Math.max(1, (maxValue - minValue) * 0.18);
    minValue -= margin;
    maxValue += margin;

    const x = i => padding + (width - padding * 2) * (i / (points.length - 1));
    const y = v => height - padding - (height - padding * 2) * ((v - minValue) / (maxValue - minValue || 1));
    const line = points.map((p, i) => `${x(i)},${y(p.weight)}`).join(" ");

    const circles = points.map((p, i) => `<circle cx="${x(i)}" cy="${y(p.weight)}" r="${p.target ? 6 : 4}" fill="${p.target ? "#ff4b0b" : "#16a34a"}"></circle>`).join("");
    const labels = points.map((p, i) => `<text x="${x(i)}" y="${height - 18}" text-anchor="middle" font-size="12" fill="#667085">${p.label}</text>`).join("");

    const grid = [0, 1, 2, 3].map(row => {
      const yy = padding + row * (height - padding * 2) / 3;
      const value = maxValue - row * (maxValue - minValue) / 3;
      return `<line x1="${padding}" y1="${yy}" x2="${width - padding}" y2="${yy}" stroke="#e4e7ec"></line><text x="10" y="${yy + 4}" font-size="12" fill="#667085">${Utils.one(value)}kg</text>`;
    }).join("");

    return `<svg viewBox="0 0 ${width} ${height}" width="100%" height="100%" preserveAspectRatio="none">
      ${grid}
      <polyline points="${line}" fill="none" stroke="#16a34a" stroke-width="4" stroke-linecap="round" stroke-linejoin="round"></polyline>
      <polygon points="${line} ${x(points.length - 1)},${height - padding} ${padding},${height - padding}" fill="rgba(22,163,74,.10)"></polygon>
      ${circles}
      ${labels}
    </svg>`;
  }
}
"""
```

## JavaScript OOP - `DietMatchScore`
Menghitung skor kecocokan diet berdasarkan target BMI, kalori, selisih berat, dan intensitas.

```python
JS_DIET_MATCH_SCORE = r"""
class DietMatchScore {
  constructor(result) {
    this.result = result;
    this.score = 100;
    this.good = [];
    this.bad = [];
    this.warnings = [];
    this.healthyMinWeight = 18.5 * Math.pow(result.heightM, 2);
    this.healthyMaxWeight = 24.9 * Math.pow(result.heightM, 2);
  }

  evaluateTargetBMI() {
    if (this.result.targetBMI < 18.5) {
      this.score -= this.result.targetBMI < 17 ? 65 : 45;
      this.bad.push(`Target berat membuat BMI target masuk kategori kurus. Untuk tinggi ${this.result.height} cm, target sehat minimal sekitar ${Utils.one(this.healthyMinWeight)} kg.`);
      this.warnings.push(`Target berat terlalu rendah. Pertimbangkan target minimal sekitar ${Utils.one(this.healthyMinWeight)} kg agar tetap masuk rentang BMI normal.`);
    } else if (this.result.targetBMI >= 30) {
      this.score -= 28;
      this.bad.push("Target berat masih membuat BMI target masuk kategori obesitas. Target bisa dibuat bertahap menuju rentang yang lebih sehat.");
    } else {
      this.good.push("BMI target masih berada pada rentang yang lebih aman untuk dijadikan tujuan awal.");
    }
  }

  evaluateCalories() {
    if (this.result.targetCalories < this.result.bmr) {
      this.score -= this.result.targetCalories < this.result.bmr * 0.9 ? 45 : 32;
      this.bad.push("Target kalori berada di bawah BMR. Makan di bawah BMR terlalu lama bisa membuat program sulit dijalani.");
      this.warnings.push("Target kalori di bawah BMR. Pilih intensitas lebih perlahan atau ubah target berat agar lebih realistis.");
    } else {
      this.good.push("Target kalori tidak berada di bawah BMR, sehingga lebih aman untuk panduan harian.");
    }
  }

  evaluateWeightChange() {
    const changePercent = this.result.absDiff / this.result.weight;

    if (changePercent > 0.30) {
      this.score -= 28;
      this.bad.push("Selisih target berat lebih dari 30% dari berat sekarang. Ini terlalu besar untuk satu tahap program.");
      this.warnings.push("Target perubahan berat terlalu ekstrem. Buat target bertahap, misalnya 5-10% dari berat saat ini terlebih dahulu.");
    } else if (changePercent > 0.20) {
      this.score -= 18;
      this.bad.push("Selisih target berat lebih dari 20% dari berat sekarang. Sebaiknya program dibuat bertahap.");
    } else if (changePercent > 0.10) {
      this.score -= 8;
      this.good.push("Target cukup menantang, tetapi masih bisa dipantau bertahap.");
    } else {
      this.good.push("Selisih target berat tidak terlalu besar, sehingga lebih mudah dipantau.");
    }
  }

  evaluateDirection() {
    if (this.result.direction === "loss" && this.result.bmi < 18.5) {
      this.score -= 35;
      this.bad.push("BMI saat ini sudah kurus, sehingga menurunkan berat badan tidak disarankan.");
    }

    if (this.result.direction === "gain" && this.result.bmi >= 30) {
      this.score -= 20;
      this.bad.push("BMI saat ini sudah tinggi, sehingga menaikkan berat badan perlu dipertimbangkan ulang.");
    }

    if ((this.result.intensity === "fast" || this.result.intensity === "gainFast") && (this.result.absDiff / this.result.weight) > 0.15) {
      this.score -= 8;
      this.bad.push("Program cepat untuk target besar cenderung lebih berat. Opsi perlahan biasanya lebih aman.");
    }
  }

  getStatus() {
    if (this.score >= 85) return {label: "Sangat cocok", cls: "green", color: "var(--green)", message: "Rencana ini masuk akal dan aman sebagai panduan awal."};
    if (this.score >= 70) return {label: "Cocok", cls: "blue", color: "var(--blue)", message: "Rencana cukup baik, tetapi tetap perlu dipantau."};
    if (this.score >= 50) return {label: "Perlu penyesuaian", cls: "yellow", color: "var(--yellow)", message: "Ada beberapa hal yang sebaiknya diperbaiki agar lebih aman."};
    return {label: "Tidak disarankan", cls: "red", color: "var(--red)", message: "Target atau kalori terlalu berisiko untuk dijadikan panduan."};
  }

  calculate() {
    this.evaluateTargetBMI();
    this.evaluateCalories();
    this.evaluateWeightChange();
    this.evaluateDirection();

    if (this.result.targetBMI < 16 || (this.result.targetBMI < 18.5 && this.result.targetCalories < this.result.bmr)) {
      this.score = Math.min(this.score, 20);
    }

    this.score = Math.round(Utils.clamp(this.score, 5, 100));

    return {
      score: this.score,
      good: this.good,
      bad: this.bad,
      warnings: this.warnings,
      healthyMinWeight: this.healthyMinWeight,
      healthyMaxWeight: this.healthyMaxWeight,
      status: this.getStatus()
    };
  }
}
"""
```

## JavaScript OOP - `Utils`
Berisi fungsi bantu seperti format angka dan kategori BMI.

```python
JS_UTILS = r"""
class Utils {
  static fmt(number) {
    return Math.round(Number(number)).toLocaleString("id-ID");
  }

  static one(number) {
    return Number(number).toFixed(1);
  }

  static clamp(number, min, max) {
    return Math.max(min, Math.min(max, number));
  }

  static bmiCategory(bmi) {
    if (bmi < 18.5) return {label: "Kurus", cls: "blue"};
    if (bmi < 25) return {label: "Normal", cls: "green"};
    if (bmi < 30) return {label: "Gemuk", cls: "yellow"};
    return {label: "Obesitas", cls: "red"};
  }
}
"""
```

## JavaScript OOP - `StorageManager`
Mengatur penyimpanan hasil dan jurnal berat badan di localStorage browser.

```python
JS_STORAGE_MANAGER = r"""
class StorageManager {
  constructor() {
    this.resultKey = "fitlogic_oop_result";
    this.weightLogsKey = "fitlogic_oop_weight_logs";
  }

  saveResult(result) {
    localStorage.setItem(this.resultKey, JSON.stringify(result));
  }

  loadResult() {
    try {
      const raw = localStorage.getItem(this.resultKey);
      return raw ? JSON.parse(raw) : null;
    } catch (error) {
      return null;
    }
  }

  loadWeightLogs() {
    try {
      return JSON.parse(localStorage.getItem(this.weightLogsKey) || "[]");
    } catch (error) {
      return [];
    }
  }

  addWeightLog(log) {
    const logs = this.loadWeightLogs();
    logs.unshift(log);
    localStorage.setItem(this.weightLogsKey, JSON.stringify(logs.slice(0, 30)));
  }

  deleteWeightLog(index) {
    const logs = this.loadWeightLogs();
    logs.splice(index, 1);
    localStorage.setItem(this.weightLogsKey, JSON.stringify(logs));
  }
}
"""
```

```python
JS_DASHBOARD = r"""
class Dashboard {
  constructor(app) {
    this.app = app;
  }

  macroRow(label, grams, calories, total, color, explain) {
    const pct = total > 0 ? Utils.clamp((calories / total) * 100, 0, 100) : 0;
    return `
      <div class="macro-row">
        <div class="macro-top"><span>${label}</span><span>${Utils.fmt(grams)}g (${Utils.fmt(calories)} kkal)</span></div>
        <div class="bar"><div class="fill ${color}" style="width:${pct}%"></div></div>
        <div class="help">${explain}</div>
      </div>`;
  }

  actionAdvice(result) {
    if (result.score < 50) {
      if (result.targetBMI < 18.5) return "Target terlalu rendah untuk tinggi badanmu. Naikkan target berat ke rentang normal dulu, lalu gunakan program perlahan.";
      if (result.targetCalories < result.bmr) return "Target kalori berada di bawah BMR. Pilih intensitas lebih perlahan atau buat target berat lebih dekat.";
      return "Ubah target menjadi lebih bertahap, lalu hitung ulang.";
    }

    if (result.direction === "loss") return "Jaga defisit kalori secara konsisten, prioritaskan protein, kurangi minuman manis/gorengan, dan lakukan latihan ringan 3-4 kali seminggu.";
    if (result.direction === "gain") return "Tambah kalori dari makanan padat nutrisi, prioritaskan protein, dan lakukan latihan beban agar kenaikan lebih banyak menjadi massa otot.";
    return "Jaga pola makan mendekati TDEE, timbang berat badan rutin, dan pertahankan aktivitas fisik.";
  }

  renderResult(result) {
    const area = document.getElementById("resultArea");
    const warningHtml = result.warnings.length
      ? `<div class="alert"><b>⚠️ Peringatan Target</b>${result.warnings.join("<br>")}</div>`
      : result.score >= 70
        ? `<div class="alert ok"><b>✅ Target cukup masuk akal</b>Target berat dan target kalori masih bisa dijadikan panduan awal.</div>`
        : `<div class="alert warn"><b>⚠️ Perlu dicek ulang</b>Ada beberapa hal yang perlu disesuaikan agar program lebih nyaman dijalani.</div>`;

    const goodHtml = result.good.map(item => `<div class="reason good">✅ ${item}</div>`).join("");
    const badHtml = result.bad.map(item => `<div class="reason bad">⚠️ ${item}</div>`).join("");
    const action = this.actionAdvice(result);

    const breakfast = Math.round(result.targetCalories * 0.25);
    const lunch = Math.round(result.targetCalories * 0.35);
    const dinner = Math.round(result.targetCalories * 0.30);
    const snack = Math.round(result.targetCalories * 0.10);

    area.className = "";
    area.innerHTML = `
      ${warningHtml}
      <div class="result-hero">
        <div><div class="username">${result.name}</div><span class="pill goal">${result.programLabel}</span></div>
        <div><div class="calories">${Utils.fmt(result.targetCalories)}</div><b>kkal / hari</b><br><small>Target Kalori</small></div>
      </div>

      <div class="grid grid-4">
        <div class="card metric"><div class="label">BMI Sekarang</div><div class="value">${Utils.one(result.bmi)}</div><span class="pill ${result.bmiCat.cls}">${result.bmiCat.label}</span><div class="desc">Kategori berat badan saat ini. Ini indikator awal, bukan diagnosis medis.</div></div>
        <div class="card metric"><div class="label">BMR</div><div class="value">${Utils.fmt(result.bmr)}</div><span class="unit">kkal/hari istirahat</span><div class="desc">Kalori minimum untuk fungsi dasar tubuh saat istirahat.</div></div>
        <div class="card metric"><div class="label">TDEE</div><div class="value">${Utils.fmt(result.tdee)}</div><span class="unit">kkal/hari aktif</span><div class="desc">Kalori untuk mempertahankan berat dengan aktivitasmu.</div></div>
        <div class="card metric" style="border-color:var(--primary)"><div class="label">Target Kalori</div><div class="value">${Utils.fmt(result.targetCalories)}</div><span class="unit">kkal/hari</span><div class="desc">Angka makan harian sesuai target berat dan intensitas program.</div></div>
      </div>

      <div class="grid grid-2" style="margin-top:22px">
        <div class="card shadow">
          <h2>✨ Kesimpulan & Arahan Diet</h2>
          <p class="help">Halo <b>${result.name}</b>, BMI kamu saat ini <b>${Utils.one(result.bmi)}</b> dan termasuk kategori <b>${result.bmiCat.label}</b>.</p>
          <p class="help">Target kamu adalah <b>${result.directionText}</b> dari <b>${Utils.one(result.weight)} kg</b> ke <b>${Utils.one(result.targetWeight)} kg</b>. BMI target menjadi <b>${Utils.one(result.targetBMI)}</b> (${result.targetBmiCat.label}).</p>
          <p class="help">Target kalori harian yang disarankan sistem adalah <b>${Utils.fmt(result.targetCalories)} kkal</b>. Estimasi waktu jika konsisten sekitar <b>${result.estimateWeeks} minggu</b>.</p>
          <p class="help"><b>Arahan aksi:</b> ${action}</p>
        </div>

        <div class="card score-card shadow">
          <h2>♡ Diet Match Score</h2>
          <div class="score-ring" style="--score:${result.score};--color:${result.status.color}">
            <div class="score-inner">${result.score}%</div>
          </div>
          <span class="pill ${result.status.cls}">${result.status.label}</span>
          <p class="help">${result.status.message}</p>
          <p class="help"><b>Catatan:</b> skor ini bukan nilai kesehatan tubuh, tapi nilai kecocokan rencana diet.</p>
        </div>
      </div>

      <div class="card shadow" style="margin-top:22px">
        <h2>Kenapa skor Diet Match seperti itu?</h2>
        <p class="help">Sistem mengecek 4 hal: BMI target, target kalori dibanding BMR, besar perubahan berat, dan intensitas program.</p>
        ${goodHtml}
        ${badHtml || '<div class="reason good">✅ Tidak ada masalah besar pada target.</div>'}
      </div>

      <div class="card shadow" style="margin-top:22px">
        <h2>Pembagian Makronutrisi</h2>
        ${this.macroRow("Protein", result.protein, result.proteinCal, result.targetCalories, "green", "Fungsi: membantu menjaga dan membangun otot. Usahakan ada lauk protein di setiap makan besar.")}
        ${this.macroRow("Karbohidrat", result.carb, result.carbCal, result.targetCalories, "orange", "Fungsi: sumber energi utama. Bisa dari nasi, kentang, roti, oat, ubi, atau buah.")}
        ${this.macroRow("Lemak", result.fat, result.fatCal, result.targetCalories, "purple", "Fungsi: bantu hormon dan metabolisme. Sumbernya bisa telur, ikan, alpukat, kacang, minyak secukupnya.")}
        <div class="macro-edu">
          <div class="macro-card"><h3>🥩 Protein</h3><p class="help">Contoh: ayam, ikan, telur, tahu, tempe, daging tanpa lemak, susu/yogurt.</p></div>
          <div class="macro-card"><h3>🍚 Karbo</h3><p class="help">Contoh: nasi, kentang, ubi, oat, roti gandum, pasta, buah.</p></div>
          <div class="macro-card"><h3>🥑 Lemak</h3><p class="help">Contoh: kacang, alpukat, kuning telur, ikan berlemak, minyak zaitun secukupnya.</p></div>
        </div>
      </div>

      <div class="card shadow" style="margin-top:22px">
        <h2>🍽️ Contoh Pembagian Menu Harian</h2>
        <div class="grid grid-4">
          <div class="menu-card"><b>🍳 Sarapan</b><span class="kcal">${breakfast} kkal</span><p>Nasi/oat/roti + telur/ayam/tahu + buah.</p></div>
          <div class="menu-card"><b>🍛 Makan Siang</b><span class="kcal">${lunch} kkal</span><p>Nasi/umbi + lauk protein + sayur + sedikit lemak sehat.</p></div>
          <div class="menu-card"><b>🌙 Makan Malam</b><span class="kcal">${dinner} kkal</span><p>Protein + karbohidrat secukupnya + sayur.</p></div>
          <div class="menu-card"><b>🍎 Camilan</b><span class="kcal">${snack} kkal</span><p>Buah, yogurt, susu, kacang, atau camilan sehat lain.</p></div>
        </div>
      </div>

      <div class="center" style="margin-top:28px">
        <button class="btn" onclick="openPage('dash')">Lihat Dashboard Aksi</button>
        <button class="btn secondary" onclick="openPage('calc')">Hitung Ulang</button>
      </div>`;
  }

  renderDashboard(result) {
    const area = document.getElementById("dashArea");
    const logs = this.app.storage.loadWeightLogs();
    const donePct = DietProgress.calculateProgressPercent(result, logs);
    const safeText = result.score >= 70 ? "Aman untuk panduan awal" : result.score >= 50 ? "Perlu dicek ulang" : "Tidak disarankan";
    const maintDiff = Math.round(result.targetCalories - result.tdee);
    const svg = DietProgress.makeTrendSvg(result, logs);
    const timeline = DietProgress.makeTimeline(result);

    area.className = "";
    area.innerHTML = `
      <div class="stat-strip">
        <div class="stat"><small>Keputusan Utama</small><b>${safeText}</b><p class="help">${result.status.message}</p></div>
        <div class="stat"><small>Target Berat</small><b>${Utils.one(result.weight)} → ${Utils.one(result.targetWeight)} kg</b><p class="help">${result.directionText}</p></div>
        <div class="stat"><small>Target Kalori</small><b>${Utils.fmt(result.targetCalories)} kkal</b><p class="help">${maintDiff >= 0 ? "+" : ""}${Utils.fmt(maintDiff)} dari TDEE</p></div>
        <div class="stat"><small>Estimasi</small><b>${result.estimateWeeks} minggu</b><p class="help">perkiraan jika konsisten</p></div>
      </div>

      <div class="card shadow" style="margin-bottom:22px">
        <div class="chart-title-row">
          <div><h2>Ringkasan Target Berat</h2><p class="help">Bagian ini menunjukkan arah perubahan berat badan. Grafik akan memakai jurnal berat jika sudah ada, kalau belum maka menampilkan simulasi target.</p></div>
          <span class="pill ${result.status.cls}">Diet Match ${result.score}%</span>
        </div>
        <div class="progress-track"><div class="fill ${result.direction === "gain" ? "blue" : "green"}" style="width:${donePct}%"></div></div>
        <p class="help">Progress menuju target: <b>${donePct}%</b>. Catat berat badan di bawah agar progress lebih nyata.</p>
        <div class="svgbox">${svg}</div>
      </div>

      <div class="grid grid-2-even">
        <div class="card shadow">
          <h2>Breakdown Kalori</h2>
          <p class="help">BMR adalah batas dasar tubuh, TDEE adalah kalori maintain, sedangkan target adalah angka yang kamu ikuti.</p>
          <div class="rowline"><span>BMR</span><b>${Utils.fmt(result.bmr)} kkal</b></div>
          <div class="rowline"><span>TDEE / Maintain</span><b>${Utils.fmt(result.tdee)} kkal</b></div>
          <div class="rowline"><span>Target Harian</span><b>${Utils.fmt(result.targetCalories)} kkal</b></div>
          <div class="rowline"><span>Perubahan dari TDEE</span><b>${maintDiff >= 0 ? "+" : ""}${Utils.fmt(maintDiff)} kkal</b></div>
          <div class="reason ${result.targetCalories < result.bmr ? "bad" : "good"}">${result.targetCalories < result.bmr ? "⚠️ Target kalori di bawah BMR, lebih baik diperbaiki." : "✅ Target kalori masih di atas BMR, lebih aman untuk panduan awal."}</div>
        </div>

        <div class="card shadow">
          <h2>Makro Nutrisi Harian</h2>
          <p class="help">Jangan hafalkan angka saja. Gunakan ini sebagai komposisi piring harian.</p>
          <div class="rowline"><span>Protein</span><b>${Utils.fmt(result.protein)}g/hari</b></div>
          <div class="rowline"><span>Karbohidrat</span><b>${Utils.fmt(result.carb)}g/hari</b></div>
          <div class="rowline"><span>Lemak</span><b>${Utils.fmt(result.fat)}g/hari</b></div>
          <div class="simple-box" style="margin-top:14px"><b>Saran:</b><p class="help">Setiap makan besar usahakan ada 1 sumber protein, 1 sumber karbohidrat, dan sayur. Lemak cukup dari minyak masak, telur, kacang, ikan, atau alpukat.</p></div>
        </div>
      </div>

      <div class="card shadow" style="margin-top:22px">
        <h2>Timeline Program</h2>
        <p class="help">Timeline ini membantu melihat program sebagai perjalanan bertahap, bukan hanya angka target akhir.</p>
        <div class="timeline">${timeline}</div>
      </div>

      <div class="grid grid-2-even" style="margin-top:22px">
        <div class="card shadow">
          <h2>Checklist Harian</h2>
          <p class="help">Ini bagian yang paling praktis untuk dilakukan setelah melihat hasil.</p>
          <div class="checklist">${this.dailyChecklist(result)}</div>
        </div>

        <div class="card shadow">
          <h2>Hal yang Perlu Dipantau</h2>
          <p class="help">Kalau salah satu poin ini terasa bermasalah, target sebaiknya dihitung ulang.</p>
          <div class="checklist">
            <div class="checkitem"><span class="check">1</span><div><b>Berat badan mingguan</b><div class="help">Naik/turun terlalu cepat berarti target kalori perlu disesuaikan.</div></div></div>
            <div class="checkitem"><span class="check">2</span><div><b>Energi harian</b><div class="help">Kalau terlalu lemas, defisit mungkin terlalu besar.</div></div></div>
            <div class="checkitem"><span class="check">3</span><div><b>Konsistensi protein</b><div class="help">Protein membantu menjaga otot, terutama saat menurunkan berat badan.</div></div></div>
            <div class="checkitem"><span class="check">4</span><div><b>Aktivitas fisik</b><div class="help">Olahraga 3-4 kali seminggu lebih baik daripada hanya mengurangi makan.</div></div></div>
          </div>
        </div>
      </div>

      <div class="card shadow journal">
        <h2>📈 Jurnal Progres Berat Badan</h2>
        <p class="help">Catat berat badan secara rutin. Setelah ada beberapa data, grafik tren akan mengikuti catatan asli, bukan simulasi.</p>
        <div class="journal-form">
          <div><label>Berat Badan (kg)</label><input id="logWeight" type="number" step="0.1" value="${result.weight}"></div>
          <div><label>Tanggal</label><input id="logDate" type="date" value="${new Date().toISOString().slice(0, 10)}"></div>
          <button class="btn" onclick="addWeightLog()">Catat Berat</button>
        </div>
        <div id="journalTable"></div>
      </div>`;

    this.renderJournalTable(result);
  }

  dailyChecklist(result) {
    const items = [];
    items.push(`<div class="checkitem"><span class="check">✓</span><div><b>Makan mendekati ${Utils.fmt(result.targetCalories)} kkal</b><div class="help">Tidak harus tepat 100%, tapi usahakan mendekati target harian.</div></div></div>`);
    items.push(`<div class="checkitem"><span class="check">✓</span><div><b>Protein sekitar ${Utils.fmt(result.protein)}g/hari</b><div class="help">Bagi ke 2-4 waktu makan agar lebih mudah tercapai.</div></div></div>`);

    if (result.direction === "loss") {
      items.push(`<div class="checkitem"><span class="check">✓</span><div><b>Kurangi kalori cair dan gorengan</b><div class="help">Minuman manis sering membuat defisit gagal tanpa terasa.</div></div></div>`);
    } else if (result.direction === "gain") {
      items.push(`<div class="checkitem"><span class="check">✓</span><div><b>Tambah makanan padat nutrisi</b><div class="help">Contoh: susu, telur, kacang, alpukat, nasi, oat, ayam, tempe.</div></div></div>`);
    } else {
      items.push(`<div class="checkitem"><span class="check">✓</span><div><b>Jaga porsi stabil</b><div class="help">Fokus pada konsistensi agar berat tidak naik/turun jauh.</div></div></div>`);
    }

    items.push(`<div class="checkitem"><span class="check">✓</span><div><b>Catat berat 1-2 kali per minggu</b><div class="help">Jangan menilai dari satu hari saja, lihat tren beberapa minggu.</div></div></div>`);
    return items.join("");
  }

  renderJournalTable(result) {
    const target = document.getElementById("journalTable");
    if (!target) return;

    const logs = this.app.storage.loadWeightLogs();

    if (!logs.length) {
      target.innerHTML = '<p class="help">Belum ada jurnal. Tambahkan catatan berat badan terlebih dahulu.</p>';
      return;
    }

    const rows = logs.map((log, index) => {
      const bmi = Number(log.weight) / Math.pow(result.heightM, 2);
      const category = Utils.bmiCategory(bmi);
      return `<tr>
        <td>${log.date}</td>
        <td>${Utils.one(log.weight)} kg</td>
        <td>${Utils.one(bmi)}</td>
        <td><span class="pill ${category.cls}">${category.label}</span></td>
        <td><button class="delete" onclick="deleteLog(${index})">hapus</button></td>
      </tr>`;
    }).join("");

    target.innerHTML = `<table>
      <thead><tr><th>Tanggal</th><th>Berat</th><th>BMI</th><th>Kategori</th><th>Aksi</th></tr></thead>
      <tbody>${rows}</tbody>
    </table>`;
  }
}
"""
```

## JavaScript OOP - `FitLogicApp`
Class utama yang menghubungkan semua class JavaScript OOP.

```python
JS_FITLOGIC_APP = r"""
class FitLogicApp {
  constructor() {
    this.gender = "male";
    this.intensity = "slow";
    this.storage = new StorageManager();
    this.dashboard = new Dashboard(this);
  }

  init() {
    document.querySelectorAll(".nav button").forEach(button => {
      button.addEventListener("click", () => this.openPage(button.dataset.page));
    });

    this.refreshIntensity();

    const saved = this.storage.loadResult();
    if (saved) {
      this.dashboard.renderResult(saved);
      this.dashboard.renderDashboard(saved);
    }
  }

  openPage(page) {
    document.querySelectorAll(".page").forEach(element => element.classList.remove("active"));

    const target = document.getElementById(page);
    if (target) target.classList.add("active");

    document.querySelectorAll(".nav button").forEach(button => {
      button.classList.toggle("active", button.dataset.page === page);
    });

    window.scrollTo(0, 0);

    if (page === "result" || page === "dash") {
      const saved = this.storage.loadResult();
      if (saved) {
        this.dashboard.renderResult(saved);
        this.dashboard.renderDashboard(saved);
      }
    }
  }

  selectGender(gender) {
    this.gender = gender;
    document.querySelectorAll("[data-gender]").forEach(element => {
      element.classList.toggle("active", element.dataset.gender === gender);
    });
  }

  getNumber(id, label, min, max) {
    const value = Number(document.getElementById(id).value);

    if (!Number.isFinite(value)) throw new Error(label + " harus berupa angka.");
    if (value < min || value > max) throw new Error(label + " sebaiknya berada pada rentang " + min + " sampai " + max + ".");

    return value;
  }

  getInputUser() {
    return new User({
      name: document.getElementById("name").value.trim() || "Pengguna",
      age: this.getNumber("age", "Umur", 10, 100),
      gender: this.gender,
      height: this.getNumber("height", "Tinggi badan", 100, 230),
      weight: this.getNumber("weight", "Berat badan", 25, 250),
      targetWeight: this.getNumber("targetWeight", "Target berat badan", 25, 250),
      activity: Number(document.getElementById("activity").value)
    });
  }

  refreshIntensity() {
    const weight = Number(document.getElementById("weight").value);
    const targetWeight = Number(document.getElementById("targetWeight").value);
    const box = document.getElementById("intensityBox");
    const help = document.getElementById("goalHelp");

    if (!Number.isFinite(weight) || !Number.isFinite(targetWeight) || weight <= 0 || targetWeight <= 0) {
      box.innerHTML = "";
      help.textContent = "Isi berat sekarang dan target berat badan terlebih dahulu.";
      return;
    }

    const direction = new User({
      name: "",
      age: 20,
      gender: this.gender,
      height: 170,
      weight: weight,
      targetWeight: targetWeight,
      activity: 1.2
    }).direction;

    if (direction === "loss") {
      if (["gainSlow", "gainFast", "maintain"].includes(this.intensity)) this.intensity = "slow";

      box.innerHTML = `
        <div class="option ${this.intensity === "slow" ? "active" : ""}" onclick="selectIntensity('slow')">
          <div class="bigicon">↘</div><b>Turun Perlahan</b><small>Defisit ±250 kkal/hari</small>
        </div>
        <div class="option ${this.intensity === "fast" ? "active" : ""}" onclick="selectIntensity('fast')">
          <div class="bigicon">🔥</div><b>Turun Cepat</b><small>Defisit ±500 kkal/hari</small>
        </div>`;
      help.textContent = "Target berat lebih rendah, jadi sistem membaca tujuanmu sebagai menurunkan berat badan.";
      return;
    }

    if (direction === "gain") {
      if (["slow", "fast", "maintain"].includes(this.intensity)) this.intensity = "gainSlow";

      box.innerHTML = `
        <div class="option ${this.intensity === "gainSlow" ? "active" : ""}" onclick="selectIntensity('gainSlow')">
          <div class="bigicon">↗</div><b>Naik Perlahan</b><small>Surplus ±250 kkal/hari</small>
        </div>
        <div class="option ${this.intensity === "gainFast" ? "active" : ""}" onclick="selectIntensity('gainFast')">
          <div class="bigicon">⚡</div><b>Naik Cepat</b><small>Surplus ±400 kkal/hari</small>
        </div>`;
      help.textContent = "Target berat lebih tinggi, jadi sistem membaca tujuanmu sebagai menaikkan berat badan.";
      return;
    }

    this.intensity = "maintain";
    box.innerHTML = `
      <div class="option active" style="grid-column:1/-1" onclick="selectIntensity('maintain')">
        <div class="bigicon">⚖️</div><b>Maintain</b><small>Target kalori mendekati TDEE</small>
      </div>`;
    help.textContent = "Target berat hampir sama dengan berat sekarang, jadi sistem membaca tujuanmu sebagai mempertahankan berat badan.";
  }

  selectIntensity(value) {
    this.intensity = value;
    this.refreshIntensity();
  }

  resetForm() {
    document.getElementById("name").value = "fayadh";
    document.getElementById("age").value = 21;
    document.getElementById("height").value = 179;
    document.getElementById("weight").value = 69;
    document.getElementById("targetWeight").value = 65;
    document.getElementById("activity").value = "1.2";
    this.selectGender("male");
    this.intensity = "slow";
    this.refreshIntensity();
  }

  calculate() {
    try {
      const user = this.getInputUser();
      if (user.direction === "maintain") this.intensity = "maintain";

      const calories = new CalorieCalculator(user, this.intensity).calculate();
      const progress = new DietProgress(user, this.intensity).calculate();
      const macros = new MacroRecommendation(calories.targetCalories, user.weight, user.direction).calculate();

      const baseResult = {
        name: user.name,
        age: user.age,
        gender: user.gender,
        height: user.height,
        weight: user.weight,
        targetWeight: user.targetWeight,
        activity: user.activity,
        heightM: user.heightM,
        direction: user.direction,
        bmi: user.bmi,
        targetBMI: user.targetBMI,
        bmr: calories.bmr,
        tdee: calories.tdee,
        targetCalories: calories.targetCalories,
        delta: calories.delta,
        rate: progress.rate,
        absDiff: progress.absDiff,
        estimateWeeks: progress.estimateWeeks,
        intensity: this.intensity,
        programLabel: calories.programLabel,
        directionText: calories.directionText,
        ...macros
      };

      const scoreResult = new DietMatchScore(baseResult).calculate();

      const finalResult = {
        ...baseResult,
        ...scoreResult,
        bmiCat: Utils.bmiCategory(baseResult.bmi),
        targetBmiCat: Utils.bmiCategory(baseResult.targetBMI),
        date: new Date().toISOString().slice(0, 10)
      };

      this.storage.saveResult(finalResult);
      this.dashboard.renderResult(finalResult);
      this.dashboard.renderDashboard(finalResult);
      this.openPage("result");
    } catch (error) {
      alert(error.message);
    }
  }

  addWeightLog() {
    const result = this.storage.loadResult();
    if (!result) return;

    const weight = Number(document.getElementById("logWeight").value);
    const date = document.getElementById("logDate").value;

    if (!Number.isFinite(weight) || weight <= 0 || !date) {
      alert("Isi berat badan dan tanggal terlebih dahulu.");
      return;
    }

    this.storage.addWeightLog({date, weight});
    this.dashboard.renderDashboard(result);
  }

  deleteLog(index) {
    const result = this.storage.loadResult();
    this.storage.deleteWeightLog(index);
    if (result) this.dashboard.renderDashboard(result);
  }
}
"""
```
